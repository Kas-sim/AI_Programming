# List, Set, Tuple, Range, Dictionary

```python
my_list = ["MQ", 'A']
print(my_list)
my_list.append(69)
print(my_list)
```

```python
print(list(range(10, 0, -2)))
```

```python
person = {
    "name":"MQ",
    "age": 25,
    "balance": 69000000
}
print(person)

person["balance"] = 99999999
print(f" {person["name"]} is just {person["age"]} and he already earns {person["balance"]} every day!")
```

```python
for key in person:
    print(key, person[key])

for key, value in person.items():
    print(key, value)
```

```python
person.keys()
person.values()
person.items()
person.update({"city": "Shanghai"})
print(person)
```

# Range + Dictionary

```python
squares = {x: x*x for x in range(1,6)}
print(squares)
```

```python
# Binary wali exponentional 2^0 + 2^1 + 2^2
y = 2
binaries = {x: y**x for x in range(0,11)}
print(binaries)
```
