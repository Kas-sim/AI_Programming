```python
class BankAccount:
    def __init__(self, balance):
        self._balance = balance

    @property
    def balance(self):
        return self._balance
    
    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Nice Try.")
        self._balance += amount
```

```python
class Account():
    def withdraw(self, amount):
        self._balance -= amount
```

```python
acct = BankAccount(100)
acct.deposit(500)
print(acct.balance)
```

```python
class Engine:
    def start(self):
        print("engine started!")

class Car:
    def __init__(self):
        self.engine = Engine()
    
    def drive(self):
        self.engine.start()
```
