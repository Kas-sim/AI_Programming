```python
# TAsk 2

city_list = ["Karachi", "Sukkur", "Karachi", "Islamabad", "Hyderabad"]
countries_tuple = ("Pakistan", "Palestine", "FinLand", "India", "GreenLand")
languages_set = {"English", "Sindhi", "Urdu", "Punjabi", "Balochi"}

city_list.append("Quetta")
print("Third country: ", countries_tuple[2])
languages_set.add("Chinese")

city_set = (city_list)
languages_list = [languages_set]

print(f"List: {city_list}")
print(f"tuple: {countries_tuple}")
print(f"Set: {languages_set}")


print(f"List into Set: {city_set}")
print(f"Set into List: {languages_list}")

print(f"\n\nList is Mutable, Ordered, can be sliced/indexed and allow dups")
print(f"Tuple is not Mutable, Ordered, can be sliced/indexed and allow dups")
print(f"Set is Mutable, Un-ordered, can not be sliced/indexed and does not allow dups")
```
