```python
# Task 2
student_info = {
    "Student_ID" : [201,202,203,204,205],
    "Name" : ["Talha", "Ahmed", "Shahzaib", "Fiaz", "Akbar"],
    "Department" : ["CS", "IT", "CS", "SE", "IT"],
    "Marks" : [86,74,91,68,88],
    "Attendance" : [90,82,96,78,92]
}

df = pd.DataFrame(student_info)
IT_students = df[df["Department"] == "IT"]
print("IT Students:\n",IT_students)

avg_marks = df.groupby("Department")["Marks"].mean()
print("Average Marks of each Department:\n",avg_marks)

status_list = []
for marks in df["Marks"]:
    if marks >= 70:
        status_list.append("Pass")
    else:
        status_list.append("Fail")

df["Status"] = status_list

print("Students with Attendance less than 85%:\n", df[df["Attendance"] < 85])

print("Sorted DataFrame by Marks(highest to lowest):",df.sort_values("Marks", ascending=False))
```
