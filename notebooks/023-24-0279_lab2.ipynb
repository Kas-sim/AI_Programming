# Task 1: Student Marks


```python
students = {
    "Ali": [78, 85, 90],
    "Sara": [88, 92, 80],
    "Ahmed": [60, 65, 70],
    "Ayesha": [95, 98, 97],
    "Usman": [72, 75, 70]
}

averages = {}   
grades = {}     
```

```python
for name in students:
    marks = students[name]
    total = 0

    for m in marks:
        total += m

    avg = total / len(marks)
    averages[name] = avg

    if avg >= 90:
        grades[name] = "A"
    elif avg >= 80:
        grades[name] = "B"
    elif avg >= 70:
        grades[name] = "C"
    else:
        grades[name] = "Fail"
```

```python
top_student = None
top_average = -1

for name in averages:
    if averages[name] > top_average:
        top_average = averages[name]
        top_student = name

print("Student Averages and Grades:")
for name in students:
    print(name, "- Average:", averages[name], "Grade:", grades[name])

print("\nTop Performer:", top_student, "with average", top_average)
```

# Sensor Data Validation using NumPy

```python
import numpy as np

readings = np.array([25, -5, 40, 110, 60, 85, -10, 95])

total = 0
count = 0
```

```python

for value in readings:
    if value >= 0 and value <= 100:
        total += value
        count += 1

if count > 0:
    average = total / count
else:
    average = 0
```

```python
corrected_count = 0

for i in range(len(readings)):
    if readings[i] < 0 or readings[i] > 100:
        readings[i] = average
        corrected_count += 1

print("Corrected sensor readings:", readings)
print("Number of corrected readings:", corrected_count)
print("Average used for correction:", average)
```
